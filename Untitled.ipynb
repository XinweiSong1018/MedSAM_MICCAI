```python
from utils import *

def read_labels(root_path):
    """
    Reads labels and returns them in a tuple of sorted lists
    """
    map_1 = sorted(glob.glob(root_path + 'Maps1_T/*.png'))
    map_2 = sorted(glob.glob(root_path + 'Maps2_T/*.png'))
    map_3 = sorted(glob.glob(root_path + 'Maps3_T/*.png'))
    map_4 = sorted(glob.glob(root_path + 'Maps4_T/*.png'))
    map_5 = sorted(glob.glob(root_path + 'Maps5_T/*.png'))
    map_6 = sorted(glob.glob(root_path + 'Maps6_T/*.png'))
    return [map_1, map_2, map_3, map_4, map_5, map_6]


# fill your path from your local folder path here
root_path = './MICCAI_2019_pathology_challenge/'
path_to_save_labels = './labels'
make_dirs(path_to_save_labels)

# Read paths from 6 annotators
maps = read_labels(root_path)
img_paths = sorted(glob.glob(os.path.join(root_path,'Train Imgs/*')))
print('Imgs found:', len(img_paths))
assert len(img_paths)==244 , 'Check your path'
# Majority voting and save image label
preprocess_labels(maps, img_paths, path_to_save_labels)



```
